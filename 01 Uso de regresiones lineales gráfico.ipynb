```python
from sklearn.linear_model import LinearRegression
```

```python
import matplotlib.pyplot as plt
```

# Datos de Entrenamiento

```python
x = [[6], [8], [10], [14],   [18]]
```

```python
y = [[7], [9], [13], [17.5], [18]]
```

```python
x2 = [[0], [10], [14], [27]]
```

# Creacion del modelo de entrenamiento

```python
model = LinearRegression()
model.fit(x, y)
```

# Ejemplo de uso

```python
print ('Una pizza de 12 cm debería costar: %.2f $' % model.predict([[15]])[0][0])
```

# Inicialización

```python
plt.figure()
plt.title('Precio de una pizza segun su diametro')
plt.xlabel('Diametro en centimetros')
plt.ylabel('Precio en Pesos')
plt.axis([0, 25, 0, 25])
plt.grid(True)
y2 = model.predict(x2)
plt.plot(x, y, 'k-')
plt.plot(x2, y2, 'g-')
plt.show()
```
